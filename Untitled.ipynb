```python
import numpy as np
import pandas as pd
import os
import PIL
from PIL import Image
import imagehash
```

```python
folder_path ="C:/Users/hp/Downloads/archive/Detect_solar_dust/Dusty"
def remove_corrupt_images(folder_path):
    for root, _, files in os.walk(folder_path):
        for file in files:
            try:
                img_path = os.path.join(root, file)
                img = Image.open(img_path)
                img.verify()  # Verify image integrity
            except Exception:
                print(f"Removing corrupt file: {img_path}")
                os.remove(img_path)

remove_corrupt_images("C:/Users/hp/Downloads/archive/Detect_solar_dust")
```

```python
def resize_jpg_images(folder_path, size=(224, 224)):
    for root, _, files in os.walk(folder_path):
        for file in files:
            if not file.lower().endswith(".jpg"):
                continue  # Only work with .jpg files

            img_path = os.path.join(root, file)

            try:
                # Open image safely
                with Image.open(img_path) as img:
                    # Convert image to RGB to avoid "RGBA/P mode" issues
                    img = img.convert("RGB")
                    img = img.resize(size)
                    img.save(img_path)  # Overwrite original
            except Exception as e:
                print(f" Corrupt or invalid image removed: {img_path} | Reason: {e}")
                try:
                    os.remove(img_path)  # Delete corrupt image
                except:
                    pass

# Run this on your dataset folder
resize_jpg_images("C:/Users/hp/Downloads/archive/Detect_solar_dust")
```

```python
def remove_duplicates_across_folders(folder_path):
    hashes = set()
    for root, _, files in os.walk(folder_path):
        for file in files:
            img_path = os.path.join(root, file)
            try:
                img = Image.open(img_path)
                hash_val = imagehash.average_hash(img)
                if hash_val in hashes:
                    print(f"Duplicate found, removing: {img_path}")
                    os.remove(img_path)
                else:
                    hashes.add(hash_val)
            except:
                continue

remove_duplicates_across_folders("C:/Users/hp/Downloads/archive/Detect_solar_dust/Dusty")
remove_duplicates_across_folders("C:/Users/hp/Downloads/archive/Detect_solar_dust/Clean")
```

```python
folder_path = "C:/Users/hp/Downloads/archive/Detect_solar_dust/Clean"
def rename_images_in_sequence(folder_path, prefix="Imgclean_", extension=".jpg"):
    # Get only .jpg files
    files = [f for f in os.listdir(folder_path) if f.lower().endswith(extension)]

    # Sort alphabetically (optional)
    files.sort()

    # Rename each file one by one
    for idx, old_name in enumerate(files, start=1):
        old_path = os.path.join(folder_path, old_name)
        new_name = f"{prefix}{idx}{extension}"
        new_path = os.path.join(folder_path, new_name)

        # Rename file
        os.rename(old_path, new_path)
        print(f"Renamed: {old_name} → {new_name}")

# 📁 Set your clean folder path
clean_folder_path = "C:/Users/hp/Downloads/archive/Detect_solar_dust/Clean"

# ▶️ Run it
rename_images_in_sequence(clean_folder_path)
```
